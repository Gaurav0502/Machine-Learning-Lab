# Importing the required modules

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import RandomizedSearchCV
```

# Reading the dataset

```python
wine = pd.read_csv(r"C:\Users\mitug\Machine-Learning-Lab\Lab-11_Grid-Search_and_Random-Search\winequality-white.csv",sep=";")
```

```python
wine.info()
```

# Test-Train Split

```python
X_train,X_test,y_train,y_test = train_test_split(wine,
    wine["quality"],
    test_size = 0.10,
    train_size=0.90,
    random_state = 0
)
```

```python
X_train.pop("quality")
X_test.pop("quality")
```

# Logistic Regression Model

```python
m = LogisticRegression(C=9,penalty='l1',solver='liblinear')
m.fit(X_train,y_train)
```

```python
y_pred_train = m.predict(X_train)
y_pred_test = m.predict(X_test)
```

```python
metrics.accuracy_score(y_true=y_train,y_pred=y_pred_train)
```

```python
metrics.accuracy_score(y_true=y_test,y_pred=y_pred_test)
```

# Grid Search

```python
space = dict()
space['solver'] = ['newton-cg', 'lbfgs']
space['penalty'] = ['l1', 'l2', 'elasticnet']
space['C'] = [8,9,10]
```

# Inferences:
1. The Logistic Regression model tuned using grid search gives a better score than the one tuned using random search.
2. Addition of more hyperparameters in the search space makes grid search more time consuming than random search.
3. However, random search might not give the optimal result which grid search will definitely give.

# Multilayer Perceptron Model

```python
classifier = MLPClassifier()
classifier.fit(X_train,y_train)
```

```python
metrics.accuracy_score(y_true=y_train,y_pred=classifier.predict(X_train))
```

```python
metrics.accuracy_score(y_true=y_test,y_pred=classifier.predict(X_test))
```

# GridSearch

```python
space = dict()
space["hidden_layer_sizes"] = [(100,),(100,100,),(100,100,100,)]
space['activation'] = ['identity', 'logistic', 'tanh', 'relu']
space['solver'] = ['lbfgs', 'sgd', 'adam']
space['early_stopping'] = [True]
```

```python
clf = GridSearchCV(classifier, space)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.cv_results_
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
classifier_grid = MLPClassifier(activation="tanh",early_stopping=True,hidden_layer_sizes=(100,),solver='lbfgs',max_iter=10000)
classifier_grid.fit(X_train,y_train)
```

```python
metrics.accuracy_score(y_true=y_train,y_pred=classifier_grid.predict(X_train))
```

```python
metrics.accuracy_score(y_true=y_test,y_pred=classifier_grid.predict(X_test))
```

# Random Search 

```python
search = RandomizedSearchCV(classifier, space, scoring='accuracy')
```

```python
search.fit(X_train,y_train)
```

```python
search.cv_results_
```

```python
search.best_score_
```

```python
search.best_params_
```

```python
classifier_r = MLPClassifier(activation="relu",early_stopping=True,hidden_layer_sizes=(100,),solver='lbfgs',max_iter=10000)
classifier_r.fit(X_train,y_train)
```

```python
metrics.accuracy_score(y_true=y_train,y_pred=classifier_r.predict(X_train))
```

```python
metrics.accuracy_score(y_true=y_test,y_pred=classifier_r.predict(X_test))
```

# Inferences:
1. The random search executes faster than the grid search.
2. For MLP model, random search did not give a better accuracy than the default model.
3. However, grid search gave a better accuracy on train data but pretty low on test data.
4. max_iter=10000 is added to let the models converge.
5. The difference between train and test accuracy for the model tuned using random search is lower compared to that of the one tuned by grid search. Hence, the former one is able to generalise better than the latter one.
